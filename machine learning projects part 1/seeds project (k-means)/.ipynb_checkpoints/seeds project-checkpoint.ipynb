```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Seed_Data.csv')
```

```python
sns.lmplot('A', 'C', data=df, hue='target', fit_reg=False, size=6);
```

```python
sns.lmplot('A', 'A_Coef', data=df, hue='target', fit_reg=False, size=6, palette='rainbow');
```

```python
g = sns.FacetGrid(data=df, hue='target', size=6, aspect=2)
g = g.map(plt.hist, 'A', bins=20, alpha=0.5);
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(df.drop('target', axis=1))
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
df['klabels'] = kmeans.labels_
df.head()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.scatter(x=df['A'], y=df['A_Coef'], c=df['klabels'], cmap='rainbow')
ax1.set_title('k means (k=3)')
ax1.scatter(x=centers[:, 0], y=centers[:, 5], c='black', s=300, alpha=0.25)

ax2.scatter(x=df['A'], y=df['A_Coef'], c=df['target'], cmap='rainbow');
ax2.set_title('original')
```

```python
sum_dist = {}
for k in range(1, 10):
    k_means = KMeans(n_clusters=k).fit(df.drop('target', axis=1))
    sum_dist[k] = k_means.inertia_
```

```python
plt.plot(list(sum_dist.keys()), list(sum_dist.values()), linestyle='-', color='g', marker='H', markerfacecolor='b')
plt.show()
```
